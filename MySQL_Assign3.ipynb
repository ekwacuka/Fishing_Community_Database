```python
import mysql.connector as mysql


mydb = mysql.connect(
host = "localhost",
user = "root",
password = "emilykimani",
database = "fishing")

mycursor = mydb.cursor()

mycursor.execute("""SELECT *
    FROM (boat
    INNER JOIN fisher 
        ON boat.boat_id = fisher.boat_id)
    INNER JOIN owner
        ON fisher.boat_id = owner.boat_id""");

print(mycursor.fetchall())

mydb.commit()    
mycursor.close()


```
